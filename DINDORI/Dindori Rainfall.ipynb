```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy.signal import correlate
```

```python
df = pd.read_excel('Dindori_updated_Q_RF.xlsx')
df = df.drop(columns = ['Dates', 'Year', 'Month', 'Day', 'Grid No.', 'Grid Name', 'LAT', 'LONG', 'Weights'])
df.head()
```

```python
df['LRF'] = pd.to_numeric(df['LRF'], errors='coerce')
correlation = df.corr()
correlation
```

```python
#correlation = df.to_csv('correlation')
excelfile = correlation.to_excel('correlation_excel.xlsx')
```
