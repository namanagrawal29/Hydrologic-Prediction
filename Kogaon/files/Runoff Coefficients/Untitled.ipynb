```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Kogaon_STREAMFLOW.xlsx')
df
```

```python
start = 0
end = 364
k = 1
intervals = []
for i in range(0, 42):
    if k == i:
        interval = (start, end)
        intervals.append(interval)
        k = k + 4
        start = end + 1
        end = end + 366
    elif k != i:
        interval = (start, end)
        intervals.append(interval)
        start = end + 1
        end = end + 365

df['Q'] = pd.to_numeric(df['Q'], errors='coerce')
sums_array = []

# Loop through each interval, calculate the sum, and append it to the list
for interval in intervals:
    start_index, end_index = interval
    sum_interval = df.loc[start_index:end_index, 'Q'].sum()
    sums_array.append(sum_interval)

if len(sums_array) < len(df):
    diff = len(df) - len(sums_array)
    sums_array.extend([np.nan] * diff)

# Add the sums_array as a new column in the DataFrame
df['sums_column'] = sums_array
# Display the array with sums for each interval
print("Sums Array:", sums_array)
```

```python
df['LRF'] = pd.to_numeric(df['LRF'], errors='coerce')
sums_LRF = []

for interval in intervals:
    start_index, end_index = interval
    sum_interval = df.loc[start_index:end_index, 'LRF'].sum()
    sums_LRF.append(sum_interval)

if len(sums_LRF) < len(df):
    diff = len(df) - len(sums_LRF)
    sums_LRF.extend([np.nan] * diff)


# Add the sums_array as a new column in the DataFrame
df['SUM LRF'] = sums_LRF
# Display the array with sums for each interval
print("Sums Array:", sums_LRF)
```
