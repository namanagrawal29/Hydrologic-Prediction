```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Patan_Q_yearly.xlsx')
df.drop(columns=['Lat', 'Long', 'weights', 'Unnamed: 25','RF_1', 'RF_2', 'RF_3', 'RF_4', 'RF_5', 'RF_6', 'RF_7', 'RF_8', 'RF_9', 'RF_10', 'RF_11', 'RF_12', 'Start', 'End', 'Q Sum', 'Grid No.', 'Unnamed: 26', 'Unnamed: 27', 'Unnamed: 28', 'Unnamed: 29', 'Unnamed: 30','Unnamed: 31', 'Unnamed: 32', 'Unnamed: 33', 'Unnamed: 34', 'Unnamed: 35'], inplace=True)
df
```

```python
df.to_excel('LRF_Q_data.xlsx')
```

```python
k = 0
j = 0
intervals = []

for i in range(0, 41):
    if k == i:
        intervals.extend(range(122 + 365*i + j, 366 + 365*i + j))
        k = k + 4
        j = j + 1
    elif k != i:
        intervals.extend(range(122 + 365*i + j, 366 + 365*i + j - 1))

# Function to get the sum of values in each interval
def get_sum_in_intervals(interval_list, data_frame, column_name):
    result = []
    for interval_start in interval_list:
        interval_end = interval_start + 1
        interval_sum = data_frame.loc[(data_frame[column_name] >= interval_start) & (data_frame[column_name] < interval_end), column_name].sum()
        result.append(interval_sum)
    return result

# Calculate the sum of values in each interval
interval_sums = get_sum_in_intervals(intervals, df, 'Q')

print(interval_sums)
```
