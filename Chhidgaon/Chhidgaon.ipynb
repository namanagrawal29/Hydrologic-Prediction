```python

# Extract the streamflow and rainfall columns from the DataFrame
R1 = df['RF_Grid 1']
Q = df['RF_Grid 1']
#streamflow = data['streamflow']
#rainfall = data['rainfall']

# Calculate cross-correlation for different lag values
max_lag = 15
cross_correlations = []
for lag in range(max_lag + 1):
    # Shift the rainfall column by the lag value
    shifted_rainfall = R1.shift(lag)
    
    # Remove NaN values caused by shifting
    shifted_rainfall = shifted_rainfall.dropna()
    
    # Calculate correlation coefficient using numpy.corrcoef()
    correlation_matrix = np.corrcoef(Q[:len(shifted_rainfall)], shifted_rainfall)
    
    # Extract the correlation coefficient from the matrix
    correlation = correlation_matrix[0, 1]
    
    # Store the correlation value
    cross_correlations.append(correlation)

# Print the cross-correlation values for different lag values
for lag, correlation in enumerate(cross_correlations):
    print(f"Lag {lag}: {correlation}")
```
